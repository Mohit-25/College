```
import numpy as np

import pandas as pd
 
import matplotlib.pyplot as plt
 
import statsmodels.api as sm
 
import seaborn
 
seaborn.set()
```

```
df = pd.read_csv('diabetes2.csv')
df.head()
```

```
x1= df[['Pregnancies']]

y=df[['Outcome']]  #simple linear regression
```

```
x = sm.add_constant(x1)
 
results = sm.OLS(y,x).fit()
```

```
results.summary()
```
